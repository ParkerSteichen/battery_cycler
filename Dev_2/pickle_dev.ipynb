## write a function that will pickle the clean_dataframe and cycle_list data once processed from exported xls files
  
I know that pickle is not ideal but it will allow me to save the pandas dataframe and easily open it again when I want to plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle

import dataprep
```

```python
clean_dataframe, cycle_list = dataprep.plotprep('example_data.xls', 2)
```

```python
# now save this data into a pickle file with a function
def pickle_store(clean_dataframe, cycle_list, filename):
    """
    docstring
    """
    pickle_data = [clean_dataframe, cycle_list]
    pickle.dump(pickle_data, open('example_pickle.pickle', 'wb'))
    print('cycling data stored to {}'.format(filename))
```

```python
pickle_store(clean_dataframe, cycle_list, 'example_pickle.pickle')
```

```python
# now a simple function for unpacking that pickle data as needed for plotting functions
# I will need to update plotting functions in the future to only take a single variable instead
# of the two that they now take `clean_dataframe` and `cycle_list`

def pickle_load(filename):
    """
    docstring
    """
    clean_dataframe, cycle_list = pickle.load(open(filename, 'rb'))
    return clean_dataframe, cycle_list
```

```python
clean_dataframe, cycle_list = pickle_load('example_pickle.pickle')
```
