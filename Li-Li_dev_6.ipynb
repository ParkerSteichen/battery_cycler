```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import time

import batt_cycle
```

```python
total_start_time = time.time()
```

-----
-----
# PP-A

```python
start_time = time.time()
PP_A_data = pickle.load(open('../Data/Li-Li/Data_2_21_19/PP-A_data.pickle', 'rb'))
PP_A, PP_A_indeces = PP_A_data
PP_A['time_hours'] = PP_A['time_sec']/3600
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# plotting a cycle range
start_cycle = 1
end_cycle = 10

range_data, label = batt_cycle.cycle_range_data(PP_A, PP_A_indeces, start_cycle, end_cycle)

fig = plt.figure(figsize=(12,4))
ax = fig.add_subplot(111)
ax.set_title('Li-Li Cell (PP-A) - '+label, fontsize=16, pad=10)
ax.set_xlabel('Time ($Hours$)')
ax.set_ylabel('Potential ($V$)')
ax.set_ylim(-abs(max(range_data['voltage'], key=abs)*1.1),
         abs(max(range_data['voltage'], key=abs)*1.1))
ax.set_xlim(min(range_data['time_hours']), max(range_data['time_hours']))
ax.yaxis.grid(True)
ax.text(0.78, 0.185, '0.2 $mA$ $cm^{-2}$', transform=fig.transFigure, fontsize=12)

ax.plot(range_data['time_hours'], range_data['voltage'])

plt.show()
```
